```python
from datasets import load_dataset
```

```python
import pandas as pd
```

Функция для предобработки текстов

```python
import re
```

```python
def preprocess_text(text):
    text = re.sub(r'[\"\']', '', text)
    text = text.replace('\n', ' ')
    text = ' '.join(text.split())

    return text
```

Датасеты с суммаризациями

wiki_lingua

```python
dataset_wiki = load_dataset("wiki_lingua", name="russian", split="train")
```

```python
dataset_wiki_size = len(dataset_wiki)

wiki_text = []

for i in range(dataset_wiki_size):
    item_wiki_size = len(dataset_wiki[i]['article']['summary'])
    for j in range(item_wiki_size):
        text1 = dataset_wiki[i]['article']['summary'][j]
        text2 = dataset_wiki[i]['article']['document'][j]
        if (len(text2.split()) <= 1000 and isinstance(text1, str) and isinstance(text2, str)):
            sentence1 = preprocess_text(text1)
            sentence2 = preprocess_text(text2)
            wiki_text.append((sentence1, sentence2))
```

```python
dataset_wiki_size
```

```python
len(wiki_text)
```

```python
dataset_wiki_df = pd.DataFrame(wiki_text)
```

```python
all_strings = dataset_wiki_df[0].apply(lambda x: isinstance(x, str)).all()
print(all_strings)
```

```python
dataset_wiki_df.to_csv("sum_wiki.csv", header=False, index=False)
```

matreshka

```python
dataset_matreshka = load_dataset("zjkarina/matreshka", split="validation")
```

```python
dataset_matreshka1 = dataset_matreshka.iloc[:100].copy()
```

```python
dataset_matreshka_size = len(dataset_matreshka)

matreshka_text = []

for i in range(dataset_matreshka_size):
    text1 = dataset_matreshka[i]['summary']
    text2 = dataset_matreshka[i]['dialog']
    if (isinstance(text1, str) and text2 != None):
        text2 = ' '.join(text2)
        if (len(text2.split()) <= 1000):
            sentence1 = preprocess_text(text1)
            sentence2 = preprocess_text(text2)
            matreshka_text.append((sentence1, sentence2))
```

```python
print(dataset_matreshka_size)
print(len(matreshka_text))
```

```python
dataset_matreshka_df = pd.DataFrame(matreshka_text)
```

```python
all_strings1 = dataset_matreshka_df[0].apply(lambda x: isinstance(x, str)).all()
print(all_strings1)
all_strings2 = dataset_matreshka_df[1].apply(lambda x: isinstance(x, str)).all()
print(all_strings2)
```

```python
dataset_matreshka_df.to_csv("sum_matreshka.csv", header=False, index=False)
```

mlsum

```python
dataset_mlsum = load_dataset("mlsum", name="ru", split="train")
```

```python
dataset_mlsum_size = len(dataset_mlsum)

mlsum_text = []

for i in range(dataset_mlsum_size):
    text1 = dataset_mlsum[i]['title'] + dataset_mlsum[i]['summary']
    text2 = dataset_mlsum[i]['text']
    if (len(text2.split()) <= 1000 and isinstance(text1, str) and isinstance(text2, str)):
        sentence1 = preprocess_text(dataset_mlsum[i]['title'] + '. ' + dataset_mlsum[i]['summary'] + '.')
        sentence2 = preprocess_text(text2)
        mlsum_text.append((sentence1, sentence2))
```

```python
print(dataset_mlsum_size)
print(len(mlsum_text))
```

```python
dataset_mlsum_df = pd.DataFrame(mlsum_text)
```

```python
all_strings = dataset_mlsum_df[1].apply(lambda x: isinstance(x, str)).all()
print(all_strings)
```

```python
dataset_mlsum_df.to_csv("sum_mlsum.csv", header=False, index=False)
```

Датасеты с перефразированиями

stsb_multi_mt

```python
dataset_stsb = load_dataset("stsb_multi_mt", name="ru", split="train")
```

```python
dataset_stsb_size = len(dataset_stsb)

stsb_text = []

for i in range(dataset_stsb_size):
    if dataset_stsb[i]['similarity_score'] >= 3.5:
        sentence1 = preprocess_text(dataset_stsb[i]['sentence1'])
        sentence2 = preprocess_text(dataset_stsb[i]['sentence2'])
        stsb_text.append((sentence1, sentence2))
```

```python
dataset_stsb_df = pd.DataFrame(stsb_text)
all_strings = dataset_stsb_df[1].apply(lambda x: isinstance(x, str)).all()
print(all_strings)
```

```python
dataset_stsb_df.to_csv("texts_labeled.csv", header=False, index=False)
```

sts22-crosslingual-sts

```python
dataset_sts22 = load_dataset("mteb/sts22-crosslingual-sts", name="ru", split="test")
```

```python
dataset_sts22_size = len(dataset_sts22)

sts22_text = []

for i in range(dataset_sts22_size):
    if dataset_sts22[i]['score'] == 4:
        sentence1 = preprocess_text(dataset_sts22[i]['sentence1'])
        sentence2 = preprocess_text(dataset_sts22[i]['sentence2'])
        sts22_text.append((sentence1, sentence2))
```

```python
dataset_sts22_df = pd.DataFrame(sts22_text)
all_strings = dataset_sts22_df[1].apply(lambda x: isinstance(x, str)).all()
print(all_strings)
```

```python
dataset_sts22_df.to_csv("texts_labeled.csv", mode='a', header=False, index=False)
```

ru_paraphraser

```python
dataset_ru_par = load_dataset("merionum/ru_paraphraser", split="train")
```

```python
dataset_ru_par_size = len(dataset_ru_par)

ru_par_text = []

for i in range(dataset_ru_par_size):
    if dataset_ru_par[i]['class'] == '1':
        sentence1 = preprocess_text(dataset_ru_par[i]['text_1'])
        sentence2 = preprocess_text(dataset_ru_par[i]['text_2'])
        ru_par_text.append((sentence1, sentence2))
```

```python
dataset_ru_par_df = pd.DataFrame(ru_par_text)
```

```python
all_strings = dataset_ru_par_df[1].apply(lambda x: isinstance(x, str)).all()
print(all_strings)
```
